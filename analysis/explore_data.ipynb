# Bitcoin Price Analysis

```python
import pandas as pd
import numpy as np
# Using plotly.express
import plotly.express as px
# Using graph_objects
import plotly.graph_objects as go
```

## Read in the data

```python
df = pd.read_csv('~/projects/crypfor/data/btc_prices.csv')
```

## Review the Price over time...

```python
fig = go.Figure([go.Scatter(x=df['dt'], y=df['price'])])
fig.show()
```

### We can see that the natural line is quite hard to understand the changes in scale overtime. Lets try Log Transforming this

```python
df['pricelog'] = np.log(df['price'])
```

```python
fig = go.Figure([go.Scatter(x=df['dt'], y=df['pricelog'])])
fig.show()
```
